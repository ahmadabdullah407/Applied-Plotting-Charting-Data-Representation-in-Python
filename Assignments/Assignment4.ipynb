```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
# Loading and cleaning data
rural=pd.read_csv("dataset/pakistan-rural-population.csv",header=0,skiprows=range(0,16))
rural=rural.set_index('date')
urban=pd.read_csv("dataset/pakistan-urban-population.csv",header=0,skiprows=range(0,16))
urban=urban.set_index('date')
df=pd.merge(urban, rural,how='left', left_index=True, right_index=True)
df.columns=df.columns.str.strip()
new_df=df[['% of Total_x','% of Total_y']]
new_df.index=pd.to_datetime(new_df.index)
new_df.columns=['Urban Population','Rural Population']
# Plotting
ax = new_df.plot.area(alpha=0.3,color=['blue','grey'])
# Decorating Plot
sns.despine()
ax.set_ylabel('% Population')
ax.set_xlabel('Year')
ax.set_title('Migration from Rural Areas to Urban Areas in Pakistan')
```
