### Note( \*args and \*\*kwargs):
args and kwargs allow you to pass in variable number of arguments in your function. they are optional parameters by default so if you pass none that would not give error as long as function can handle 0 arguments. Also arguments are put into a tuple and that tuple is passed into the function in case of args and in case of kwargs they are put into a dictionary with names as keys and values associated with names as values(so basically it is for named arguments).also you do not have to exactly name parameters as args and kwargs, it is \* and \*\* that defines args and kwargs.

```python
i=1
def hello(*args,**kwargs):
    global i
    print(args)
    print(kwargs)
    print('-----------{}------------'.format(i))
    i+=1
a=(1,2,3)
b={'c':1,'d':2,'f':3}
hello(a)
hello(b)
hello(1,2,3)
hello(c=1,d=2)
hello(b=b)
hello(*a)
hello(**b)
hello(*b)
hello(*a,**b)
```

## plotting without using scripting layer:
Let's see how to make a plot without using the scripting layer, and directly interface with artist layer.

First let's set the backend without using mpl.use() from the scripting layer

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
# create a new figure
fig=Figure()
# associate fig with the backend
canvas = FigureCanvasAgg(fig)
# add a subplot to the fig
ax = fig.add_subplot(111)
# plot the point (3,2)
ax.plot(3, 2, '.')
# save the figure to test.png
# you can see this figure in your Jupyter workspace afterwards in your pc.
canvas.print_png('test.png')
```

### gca and gcf:
ok lets create a new figure with pyplot, this means it will create a new figure and would not update figure above. with gca function we can get current axis and gcf we can get current figure.

```python
plt.figure()
# plot the point (3,2) using the circle marker
plt.plot(3, 2, 'o')
# get the current axes
ax = plt.gca()
# Set axis properties [xmin, xmax, ymin, ymax]
ax.axis([0,6,0,10])
```

We can make subsequent calls to the plot function. This will add more data to our chart,You can see when this is done points are rendered in different colors as axes recognizes them as data series.

```python
# create a new figure
plt.figure()

# plot the point (1.5, 1.5) using the circle marker
plt.plot(1.5, 1.5, 'o')
# plot the point (2, 2) using the circle marker
plt.plot(2, 2, 'o')
# plot the point (2.5, 2.5) using the circle marker
plt.plot(2.5, 2.5, 'o')
```

We can go further with the axes object to the point where we can get all chuld objects that axes contains, Here we can see there are actually 
- 3 line 2d objects contained in the axes whuch represent data points.
- 4 spine objects which are actually borders of the frame including tic markers.
- 2 axis objects.
- A bunch of texts which are label of the charts.
- There is a rectangle too whicj is background of the axes.

```python
# get current axes
ax = plt.gca()
# get all the child objects the axes contains
ax.get_children()
```
