# Heatmaps:
Heatmaps are basically two dimensional histograms.

In Heatmaps you have two dimensions latitude and longitude and the we can overlay on top of it a 3rd dimnsion say temperature or a rain etc. and use color to indicate its intensity.

```python
plt.figure()

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
_ = plt.hist2d(X, Y, bins=25)
```

```python
plt.figure()
_ = plt.hist2d(X, Y, bins=100)
```
