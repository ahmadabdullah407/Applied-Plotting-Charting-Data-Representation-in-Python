# Animations:

## FuncAnimation():
It is a function that calls some other function we define iteratively to build an animation.

Now 1st lets create a numpy array of normal distribution with size 100. np.random.randn creates a normal distribution array of size specified with mean 0 and variance 1.

```python
import matplotlib.animation as animation
n = 100
x = np.random.randn(n)
x
```

Now lets define a function which should be passed within FuncAnimation function to create an animation.

```python
def update(curr):# curr is the index the function is at in array
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == n: 
        a.event_source.stop()
    # clear the axis
    plt.cla()
    # We can also passin the spacing between bins as np array
    bins = np.arange(-4, 4, 0.5)
    # Create histogram till curr index
    plt.hist(x[:curr], bins=bins)
    # we need to set axis each time bcz histogram will try to autoscale
    plt.axis([-4,4,0,30])
    # labeling
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
#     annotate places a text at specified position in the chart,
# we'll use it to show how many samples are currently being rendered to the screen
    plt.annotate('n = {}'.format(curr), [3,27])
```

in animation.FuncAnimation() we pass in figure, function and interval that is time interval between function execution in this case it is 100ms.

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100)
```
